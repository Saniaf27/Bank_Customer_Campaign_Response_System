# IMPORT LIBRARIES


```python
import csv
import os
```

# Initialize CSV Files With Headers

```python
def initialize_files():

    files={
        "customers.csv":["CustomerID","Name","Age","Job","MaritalStatus","Education"],
        "transactions.csv":["CustomerID","Balance","LastTransaction","TransactionType"],
        "credit_history.csv":["CustomerID","PreviousLoans","RepaymentHistory","MissedPayments"],
        "campaign_results.csv":["CustomerID","Prediction"],
        "support_requests.csv":["RequestID","CustomerID","Issue","Status"]
    }

    for file,fields in files.items():

        if not os.path.exists(file):

            with open(file,'w',newline='') as f:
                writer=csv.DictWriter(f,fieldnames=fields)
                writer.writeheader()
```

# Read CSV Function

```python
def read_csv(file):

    data=[]

    try:
        with open(file,'r') as f:

            reader=csv.DictReader(f)

            for row in reader:
                data.append(row)

    except FileNotFoundError:
        print("File not found:",file)

    return data
```

# CSV Function

```python
def append_csv(file,fields,data):

    with open(file,'a',newline='') as f:

        writer=csv.DictWriter(f,fieldnames=fields)
        writer.writerow(data)
```

# Manager: Add Customer

```python
def add_customer():

    data={
        "CustomerID":input("Customer ID: "),
        "Name":input("Name: "),
        "Age":input("Age: "),
        "Job":input("Job: "),
        "MaritalStatus":input("Marital Status: "),
        "Education":input("Education: ")
    }

    fields=["CustomerID","Name","Age","Job","MaritalStatus","Education"]

    append_csv("customers.csv",fields,data)

    print("Customer added successfully")
```

# Manager: Record Transaction

```python
def record_transaction():

    data={
        "CustomerID":input("Customer ID: "),
        "Balance":input("Balance: "),
        "LastTransaction":input("Transaction Amount: "),
        "TransactionType":input("Deposit/Withdrawal: ")
    }

    fields=["CustomerID","Balance","LastTransaction","TransactionType"]

    append_csv("transactions.csv",fields,data)

    print("Transaction recorded")
```

# Manager: Add Credit History

```python
def update_credit_history():

    data={
        "CustomerID":input("Customer ID: "),
        "PreviousLoans":input("Previous Loans: "),
        "RepaymentHistory":input("Repayment History (Good/Average/Poor): "),
        "MissedPayments":input("Missed Payments: ")
    }

    fields=["CustomerID","PreviousLoans","RepaymentHistory","MissedPayments"]

    append_csv("credit_history.csv",fields,data)

    print("Credit history updated")
```

# Calculate Credit Score

```python
def calculate_credit_score(cid):

    history=read_csv("credit_history.csv")
    transactions=read_csv("transactions.csv")

    score=600

    for row in history:

        if row["CustomerID"]==cid:

            if row["RepaymentHistory"].lower()=="good":
                score+=80

            if int(row["MissedPayments"])>0:
                score-=100

            if int(row["PreviousLoans"])>2:
                score-=50

    for t in transactions:

        if t["CustomerID"]==cid:

            if int(t["Balance"])>3000:
                score+=40

    return score
```

# Campaign Analysis

```python
def campaign_analysis():

    customers=read_csv("customers.csv")
    transactions=read_csv("transactions.csv")

    for c in customers:

        cid=c["CustomerID"]
        score=calculate_credit_score(cid)

        balance=0

        for t in transactions:

            if t["CustomerID"]==cid:
                balance=int(t["Balance"])

        if score>680 and balance>2500:

            prediction="Likely Subscriber"

        else:

            prediction="Not Likely Subscriber"

        data={
            "CustomerID":cid,
            "Prediction":prediction
        }

        append_csv("campaign_results.csv",
                   ["CustomerID","Prediction"],
                   data)

        print(cid,prediction)
```

# Customer: View Profile

```python
def view_profile(cid):

    customers=read_csv("customers.csv")

    for c in customers:

        if c["CustomerID"]==cid:

            print("\nCustomer Profile")
            print(c)
```

# Customer: Check Balance

```python
def check_balance(cid):

    transactions=read_csv("transactions.csv")

    for t in transactions:

        if t["CustomerID"]==cid:

            print("Balance:",t["Balance"])
```

# Customer: Raise Support Request

```python
def raise_support():

    data={
        "RequestID":input("Request ID: "),
        "CustomerID":input("Customer ID: "),
        "Issue":input("Issue: "),
        "Status":"Open"
    }

    fields=["RequestID","CustomerID","Issue","Status"]

    append_csv("support_requests.csv",fields,data)

    print("Support request submitted")
```

# Manager: View Support Requests

```python
def view_support_requests():

    requests=read_csv("support_requests.csv")

    for r in requests:

        print(r)
```

# Main Menu

```python
def main_menu():

    initialize_files()

    while True:

        print("\n===== BANK CUSTOMER CAMPAIGN RESPONSE SYSTEM =====")
        print("1 Customer Portal")
        print("2 Bank Manager Portal")
        print("3 Exit")

        choice=input("Enter choice: ")

        if choice=="1":

            cid=input("Enter Customer ID: ")

            while True:

                print("\n--- CUSTOMER MENU ---")
                print("1 View Profile")
                print("2 Check Balance")
                print("3 Check Credit Score")
                print("4 Raise Support Request")
                print("5 Back")

                c=input("Choice: ")

                if c=="1":
                    view_profile(cid)

                elif c=="2":
                    check_balance(cid)

                elif c=="3":
                    score=calculate_credit_score(cid)
                    print("Credit Score:",score)

                elif c=="4":
                    raise_support()

                elif c=="5":
                    break

                else:
                    print("Invalid option")

        elif choice=="2":

            while True:

                print("\n--- BANK MANAGER MENU ---")
                print("1 Add Customer")
                print("2 Record Transaction")
                print("3 Update Credit History")
                print("4 Run Campaign Analysis")
                print("5 View Support Requests")
                print("6 Back")

                m=input("Choice: ")

                if m=="1":
                    add_customer()

                elif m=="2":
                    record_transaction()

                elif m=="3":
                    update_credit_history()

                elif m=="4":
                    campaign_analysis()

                elif m=="5":
                    view_support_requests()

                elif m=="6":
                    break

                else:
                    print("Invalid option")

        elif choice=="3":

            print("System closed")
            break
```
